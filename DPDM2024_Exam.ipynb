## ค้นหาพยาน และคำให้การของพยานที่เห็นเหตุการณ์

```python
# prompt: สร้างตัวแปรใหม่ จาก  person_merged_data เพื่อเอา address_number ที่มีค่ามากที่สุด และน้อยที่สุด เมื่อ address_street_name คือ  Northwestern Dr โดยแสดงข้อมูลทั้งแถวในรูปแบบ dataframe

# หา address_number ที่มากที่สุด
max_address_number = person_merged_data[person_merged_data['address_street_name'] == 'Northwestern Dr']['address_number'].max()
# หา address_number ที่น้อยที่สุด
min_address_number = person_merged_data[person_merged_data['address_street_name'] == 'Northwestern Dr']['address_number'].min()
# สร้างตัวแปรใหม่
witness_11 = person_merged_data[
    (person_merged_data['address_street_name'] == 'Northwestern Dr') &
    (person_merged_data['address_number'] == max_address_number)
]
witness_12 = person_merged_data[
    (person_merged_data['address_street_name'] == 'Northwestern Dr') &
    (person_merged_data['address_number'] == min_address_number)
]
witnesses_1 = pd.concat([witness_11, witness_12])
witnesses_1
```

```python
# prompt: สร้างตัวแปรใหม่จากตัวแปร person_merged_data โดยเลือกคอลัมน์  address_street_name ที่เป็น "Franklin Ave"
# โดยให้แสดงข้อมูลที่มีเงื่อนไขดังกล่าวทั้งแถว และตั้งชื่อตัวแปรใหม่ว่า witness_2_location

witness_2_location = person_merged_data[person_merged_data['address_street_name'] == 'Franklin Ave']
witness_2 = witness_2_location[witness_2_location['name'].str.contains('Annabel')]
witness_2
```

```python
# เนื่องจาก Kinsey Erickson ไม่มีคำให้การ เราจึงไม่นำข้อมูลของเขาไปใช้ต่อ

witnesses = pd.concat([witnesses_1.iloc[0:1], witness_2])
witnesses
```

## สืบหาฆาตรจากคำให้การของพยาน

```python
# prompt: สร้างตัวแปรใหม่จากตัวแปร person_merged_data โดยเลือกคอลัมน์  membership_status ที่เป็น "gold"
# ซึ่งมี id  ขึ้นต้น 3 ตัวแรกคือ "48Z" โดยให้แสดงข้อมูลที่มีเงื่อนไขดังกล่าวทั้งแถว และตั้งชื่อตัวแปรใหม่ว่า  Suspects_membership

suspects_membership = person_merged_data[(person_merged_data['membership_status'] == 'gold') & (person_merged_data['membership_id'].str.startswith('48Z'))]
suspects_membership
```

```python
suspect_car = person_merged_data[person_merged_data['plate_number'].str.contains('H42W', na=False)]
suspect_car
```

```python
# prompt: หาข้อมูลจากตัวแปร person_merged_data โดยเลือกคอลัมน์ check_in_date ที่เป็น วันที่ 9 มกราคม

suspects_check_in = person_merged_data[person_merged_data['check_in_date'] == 20180109]
suspects_check_in
```

```python
# prompt: ทำการหาข้อมูลว่า แถวใดที่มีข้อมูลอยู่ทั้ง 3 ตัวแปร suspects_membership, suspect_car, suspects_check_in แล้วแสดงข้อมูลดังกล่าวทั้งแถว

# หา person_id ที่ซ้ำกันในทั้ง 3 ตาราง
common_person_ids = set(suspects_membership['person_id']).intersection(suspect_car['person_id']).intersection(suspects_check_in['person_id'])

# แสดงข้อมูลผู้ต้องสงสัยที่ตรงกับ person_id ที่ซ้ำกัน
suspects = person_merged_data[person_merged_data['person_id'].isin(common_person_ids)]
suspects
```
